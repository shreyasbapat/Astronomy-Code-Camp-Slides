Now, this was just an example

Let's define our own orbit

```python
from poliastro.bodies import Earth, Mars
```

```python
earth = Orbit.from_body_ephem(Earth)
mars = Orbit.from_body_ephem(Mars)
```

```python
planets = OrbitPlotter()
planets.plot(earth,label='Earth')
planets.plot(mars,label='Mars')
```

```python
planets3D = OrbitPlotter3D()
planets3D.plot(earth,label='Earth')
planets3D.plot(mars,label='Mars')
planets3D.show()
```
