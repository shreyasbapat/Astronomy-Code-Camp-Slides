```python
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
import astropy.units as u
```

```python
from poliastro.twobody.orbit import Orbit
```

```python
from poliastro.plotting import OrbitPlotter, OrbitPlotter2D, OrbitPlotter3D
```

```python
from poliastro.examples import molniya, iss
from poliastro.bodies import Sun
```

```python
hyper = Orbit.from_vectors(
    Sun,
    [ -9.77441841e+07,  1.01000539e+08,  4.37584668e+07] * u.km,
    [ 23.75936985,-43.09599568, -8.7084724 ] * u.km / u.s,
)
```

```python
a = OrbitPlotter()
a.plot(hyper)
```

```python
a_2D = OrbitPlotter2D()
a_2D.plot(hyper)
a_2D.show()
```

```python
a_3D = OrbitPlotter3D()
a_3D.plot(hyper)
a_3D.show()
```

Now, let's propagate in time :p

```python
hyper_p = hyper.propagate(25*u.d)
```

```python
a_2D_p = OrbitPlotter2D()
a_2D_p.plot(hyper_p)
a_2D_p.show()
```

We can also use some different propagation methods to propagate the orbit.
The list of methods include:
1) Mean Motion
2) Cowell
3) Kepler

### Cowell's Method of Propagation

```python
from poliastro.twobody.propagation import cowell
```

```python
hyper_p_cow = hyper.propagate(25*u.d, method=cowell)
```

```python
a_2D_p_cow = OrbitPlotter2D()
a_2D_p_cow.plot(hyper_p_cow)
a_2D_p_cow.show()
```

These differences are very minor to observe.

If you want to sample the orbit points, you can do that too!

```python
hyper.sample(values=50, method=cowell)
```
