```python
from math import degrees
sun=ephem.Sun()
ra=[]
dec=[]
for d in range(0,365):
    delhi.date = ephem.now() + d
    sun.compute(delhi)
#     print(delhi.date)
    ra.append(degrees(sun.ra))
    dec.append(degrees(sun.dec))
plt.plot(ra,dec)
plt.xlabel("ra")
plt.ylabel("dec")
plt.show()   
```
